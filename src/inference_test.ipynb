```python
import torch
from pathlib import Path

# Define the directory containing the .pth files
save_path = Path("data/saved")  # Replace with your directory path
part = "saved_bpe_02Jan/inference/test"  # Replace with the specific part directory name if applicable

# List all .pth files in the specified directory
pth_files = sorted(save_path.glob(f"{part}/output_*.pth"))

# Initialize lists to store the loaded data
predictions = []
ground_truths = []
audio_paths = []

# Loop through the .pth files and load the data
for pth_file in pth_files:
    data = torch.load(pth_file)
    predictions.append(data["prediction"])
    ground_truths.append(data["ground_truth"])
    audio_paths.append(data["audio_path"])

# Display some of the loaded data
print(f"Number of files loaded: {len(predictions)}")
print(f"First prediction: {predictions[0]}")
print(f"First ground truth: {ground_truths[0]}")
print(f"First audio path: {audio_paths[0]}")
```
